### Getting Data for EDA

```python
#using regular expressions to extract the data.

import re
def getstate(x):
  if x is not np.nan:
    st = re.search("\([a-zA-Z]*( \&)?( [a-zA-Z]*){0,3}\)",x)
    if st is not None:
      return st.group()[1:-1]

def getcity(x):
  if x is not np.nan:
    st = re.search("^[a-zA-Z]*(_| )?",x)
    if st is not None:
      return st.group()[:-1]

def city_place(x):
  #removing state
  x = x.split('(')[0] 
  len_ = len(x.split('_'))
  if len_ >= 3:
    return x.split('_')[1]
  #small cities have same city and place name
  if len_ == 2:
    return x.split('_')[0]
  #dealing with edge cases or improper naming conventions
  return x.split(' ')[0]

def citycode(x):
#removing state
  x = x.split('(')[0]
  if(len(x.split('_')) >=3):
    return x.split('_')[-1]
  return 'none'
```

```python
trip["destination_state"] = trip["destination_name"].apply(lambda x:getstate(x))
trip["destination_city"] = trip["destination_name"].apply(lambda x:getcity(x))
trip["destination_place"] = trip["destination_name"].apply(lambda x:city_place(x))
trip["destination_code"] = trip["destination_name"].apply(lambda x:citycode(x))
```

```python
trip["source_state"] = trip["source_name"].apply(lambda x: getstate(x))
trip["source_city"] = trip["source_name"].apply(lambda x: getcity(x))
trip["source_place"] = trip["source_name"].apply(lambda x: city_place(x))
trip["source_code"] = trip["source_name"].apply(lambda x: citycode(x))
```

## Hypothesis Testing

```python
1. start_scan_to_end_scan v/s od_time_diff_hour
    NUll Hypothesis:H0:The mean of both groups are equal.
    Alternate Hypothesis:H1:The mean are not equal.
    α = 0.05
```

```python
sns.distplot(trip["start_scan_to_end_scan"], label="start_scan_to_end_scan")
sns.distplot(trip["od_time_diff_hour"], label="od_time_diff_hour")
plt.legend()
plt.show()
```

- The mean seems to be almost same as per the plot. let us do 2 sample t- test to verify after normalising the data

```python
from scipy.stats import boxcox
x_trnsf1 , lambda1 = boxcox(trip["start_scan_to_end_scan"])
x_trnsf2 , lambda2 = boxcox(trip["od_time_diff_hour"])
sns.distplot(x_trnsf1)
sns.distplot(x_trnsf2)
plt.show()
```

```python
import statsmodels.api as sm
fig, ax = plt.subplots(1,2,figsize=(12,4))
sm.qqplot(x_trf1, line='s', ax = ax[0])
sm.qqplot(x_trf2, line='s', label='od_time_diff_hour', ax = ax[1])
ax[0].set_title('start_scan_to_end_scan')
ax[1].set_title('od_time_diff_hour')
plt.show()
```

As the data is still not normal even after trying BoxCox transform,we will perform a non-parametric test (Mann-Whitney).
Null Hypothesis:H0:The distributions of both samples are equal.
Alternate Hypothesis:H1:The distributions of both samples are not equal.
alpha(α)=0.05

```python
from scipy.stats import mannwhitneyu
stat, p = mannwhitneyu(trip["start_scan_to_end_scan"],trip["od_time_diff_hour"])
p
```

Since p > α,we fail to reject the null hypothesis. 
Therefore we can say that,the sample distributions seems to be the same for ‘start_scan_to_end_scan’ and ‘od_time_diff_hour’. 
So, the trip duration and the difference between trip start and end are indeed the same.

2. actual_time v/s osrm_time
    Null Hypothesis:H0:The mean of actual_time and calculated(osrm) time are equal.
    Alternate Hypothesis:H1:The mean are not equal.
    alpha(α) = 0.05

```python
sns.distplot(trip["actual_time"], label="actual")
sns.distplot(trip["osrm_time"], label="osrm")
plt.legend()
plt.show()
```

```python
from scipy.stats import boxcox
x_trnsf11 , lambda11 = boxcox(trip["actual_time"])
x_trnsf21 , lambda21 = boxcox(trip["osrm_time"])
sns.distplot(x_trnsf11)
sns.distplot(x_trnsf21)
plt.show()
```

As it is clear from the above plot that these do not follow normal distribution,we can go for the non-parametric Mann-Whitney test.

Null Hypothesis:H0:The mean of actual_time and calculated(osrm) time are equal.
Alternate Hypothesis:H1:The mean are not equal.
alpha(α) = 0.05

```python
stat, p = mannwhitneyu(trip["actual_time"].sample(1000), trip["osrm_time"].sample(1000))
p
```

Since p < α (0.05), we can reject the null hypothesis.So,can say that the actual time and the time calculated by algorithm (osrm) are very different.

3. actual_time v/s segment_actual_time
Null Hypothesis:H0 : The mean of actual_time and segment_actual_time are equal.
Alternate Hypothesis:H1: The mean are not equal.
α = 0.05

```python
sns.distplot(trip["actual_time"], label="actual")
sns.distplot(trip["segment_actual_time_total"], label="actual_segment_total")
plt.legend()
plt.show()
```

```python
from scipy.stats import boxcox
x_trnsf111 , lambda111 = boxcox(trip["actual_time"])
x_trnsf211 , lambda211 = boxcox(trip["segment_actual_time_total"])
sns.distplot(x_trnsf111)
sns.distplot(x_trnsf211)
plt.show()
```

Here we can see that the times are not normally distributed.However, the distributions look similar.So we can use the non parametric Mann-Whitney test again. 
Null Hypothesis:H0 : The median of both groups are equal.
Alternate Hypothesis:H1: The median are not equal.

```python
stat, p = mannwhitneyu(trip["actual_time"].sample(1000),trip["segment_actual_time_total"].sample(1000))
p
```

Since p > α (0.05),we fail to reject the null hypothesis.Hence, the sample distributions seem to be the same for ‘segment_actual_time’ and ‘actual_time’. .So, the actual total time taken for a trip is similar to the sum of the distances of a trip’s segments.

6. Does time taken depend on day of week?
Null Hypothesis:H0: All sample distributions are equal.
Alternate Hypothesis:H1: One or more sample distributions are not equal.

```python
plt.figure(figsize=(16,8))
days = ['Mon','Tue','Wed','Thurs','Fri','Sat','Sun']
daydist=[]
for day in days:
  daydist.append(trip[trip["trip_dayofweek"]==day]["actual_time"])
  sns.distplot(trip[trip["trip_dayofweek"]==day]["actual_time"], hist=False,label=day) 
plt.legend()
plt.show()
```

The distributions look slightly different. So,We will perform the non-parametric counterpart to one way ANOVA, the Kruskal Willis Test.

```python
stats.kruskal(daydist[0],daydist[1],daydist[2],daydist[3],daydist[4],daydist[5],daydist[6])
```

Since p >  (0.05), we fail to reject the null hypothesis. 
Hence, the sample distributions of time taken seem to be the same for all week days.
